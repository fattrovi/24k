```
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(-10,10,200)
y = np.sinc(x)+np.cos(x)
z = np.tan(y+5)
```

```
plt.plot(x,z,':g',lw=3)
```
