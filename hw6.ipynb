```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

## 1. 定義二維平面上的區域  S⊆R^2
S 為一半徑為3的圓

```
a = np.linspace(0, 3, 40)
b = np.linspace(0,2*np.pi, 80)
aa,bb = np.meshgrid(a, b)
at,bt = aa*np.cos(bb), aa*np.sin(bb)

xx, yy = at.flatten(), bt.flatten()
S_meshgrid = np.vstack([xx, yy])
```

```
plt.scatter(xx, yy, c= -aa)
plt.title('S')
plt.xlim(-4.5, 4.5)
plt.ylim(-3.2, 3.2)
```

## 2. 定義線性轉換

Φ(𝑥,𝑦)= ( 3x-5y, -5x+7y) A為一對稱矩陣

```
A = np.array([[0.6,-1],
             [-1,0.6]])
def phi(x):
    assert A.shape[1]==x.shape[0], 'incompatible dimension'
    return A@x
```

## 3. 畫  Φ(S)

```
T = phi(S_meshgrid)
u, v = phi(S_meshgrid)

plt.scatter(u, v, c= aa)
plt.title('$\Phi(S)$')
```

```
plt.scatter(xx, yy,c= 'b',alpha=0.2)
plt.scatter(u, v, c = 'r')
plt.title('$S$ and $\Phi(S)$')

plt.xlim(-5.5, 5.5)
plt.ylim(-4, 4)
```
