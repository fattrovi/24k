```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

## 1. 定義二維平面上的區域  S⊆R^2

```
x,y = 40,80
a = np.linspace(0, 3, x)
b = np.linspace(0,2*np.pi, y)
aa,bb = np.meshgrid(a, b)
at,bt = aa*np.cos(bb), aa*np.sin(bb)
```

```
plt.scatter(at, bt, c= -aa)
plt.title('S')
plt.xlim(-3.5, 3.5)
plt.ylim(-3.5, 3.5)
```
